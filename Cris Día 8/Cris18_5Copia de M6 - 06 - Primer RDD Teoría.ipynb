## Ejercicio:

Intenta leer un archivo csv o de texto y hacer alguna transformación.

* Busca el archivo del quijote.txt por internet. 
* Subelo a Google COlab e importalo con pyspark. 
* Cuenta los caracteres del fichero.

```python
# #Inserta aquí tú código

from pyspark.sql import SparkSession
from google.colab import files

# Subir el archivo "quijote.txt" a Colab
uploaded = files.upload()

# Crear una instancia de SparkSession
spark = SparkSession.builder.appName("Conteo de caracteres").getOrCreate()

# Obtener la ruta del archivo subido
file_path = next(iter(uploaded.keys()))

# Cargar el archivo de texto
lines = spark.read.text(file_path).rdd.map(lambda r: r[0])

# Contar los caracteres
character_count = lines.flatMap(lambda line: list(line)).count()

# Imprimir el resultado
print("El número de caracteres en el archivo es:", character_count)
```
